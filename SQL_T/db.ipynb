<h3 style="color: red;">CREATE DB</h3>

```python
import sqlite3
from pathlib import Path

ROOT_DIR = ROOT_DIR = Path.cwd() # Current Working Directory (Dir atual)
DB_NAME = 'dbt.sqlite3'
DB_FILE = ROOT_DIR / DB_NAME
TABLE_NAME = 'customers'

if __name__ == '__main__': 
    connection = sqlite3.connect(DB_FILE)
    cursor = connection.cursor()


    cursor.execute(
        f'''
        CREATE TABLE IF NOT EXISTS {TABLE_NAME} (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT,
            age INTEGER
        )
        '''
    )
    connection.commit()


    sql_custumers = (    
        f'''
        INSERT INTO {TABLE_NAME} (name, age)
        VALUES
            (?,?)''' #binding para previnir sql injection
    )

    dados =[
        ("Jorge", 21),
        ("Elson", 45),
        ("Rogeria", 78)
    ]

    cursor.executemany(sql_custumers, dados)
    connection.commit()


    cursor.close()
    connection.close()
```

<h3 style="color: red;">SHOW DB</h3>

```python
connection = sqlite3.connect(DB_FILE)
cursor = connection.cursor()

cursor.execute(f'SELECT * FROM {TABLE_NAME}') 


rows = cursor.fetchall() 
if not rows:
    print('No data found!')
else:
    for row in rows: 
        print(row)


cursor.close()

connection.close()
```

<h3 style="color: red;">FILTER DB</h3>

```python
connection = sqlite3.connect(DB_FILE)
cursor = connection.cursor()

cursor.execute(
    f'''SELECT NAME,AGE FROM {TABLE_NAME} 
    WHERE AGE >= 18 AND AGE <= 50'''
) 


rows = cursor.fetchall() 
if not rows:
    print('No data found!')
else:
    for row in rows:
        print(row)

cursor.close()
connection.close()

```

<h3 style="color: red;">CREATE ID DB</h3>

```python
connection = sqlite3.connect(DB_FILE)
cursor = connection.cursor()

name = input("Enter a name: ")
age = int(input("Age: "))

cursor.execute(
    f"INSERT INTO {TABLE_NAME} (name, age) VALUES (?, ?)",
    (name, age)
)

connection.commit()

print("Success!")

cursor.close()
connection.close()
```

<h3 style="color: red;">DELETE ID DB</h3>

```python
connection = sqlite3.connect(DB_FILE)
cursor = connection.cursor()

id_delete = int(input("Enter the ID: "))

cursor.execute(
    f"DELETE FROM {TABLE_NAME} WHERE id = ?",
    (id_delete,)
)

connection.commit()

print("Data deleted!")

cursor.close()
connection.close()
```

<h3 style="color: red;">* DELETE ALL IDS DB</h3>

```python
connection = sqlite3.connect(DB_FILE)
cursor = connection.cursor()


cursor.execute(
    f'DELETE FROM sqlite_sequence WHERE name="{TABLE_NAME}"' # reiniciar o autoincrement
)
connection.commit()


cursor.execute(
    f'DELETE FROM {TABLE_NAME}' # deletar a tabela toda
)

connection.commit()


print('Data deleted!')

        
cursor.close()
connection.close()
```

<h3 style="color: red;">* DELETE TABLE CUSTOMERS DB</h3>

```python
connection = sqlite3.connect(DB_FILE)
cursor = connection.cursor()

cursor.execute(
       f"DROP TABLE {TABLE_NAME}"
)

connection.commit()
```
